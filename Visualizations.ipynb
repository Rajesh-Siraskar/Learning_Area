# Visualization of data and model performance

## Error bars
- Use std. deviation to indicate uncertainty
- Error bar = std.dev values
- Use pandas to "filter" data on criteria (environment)
- Then for that environment - plot line plot with error bar -
- For a single column say A2C Precision

````python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
````

````python
# Import Data
df = pd.read_csv('R1-Concat_Experiments.csv')
print('Experiments data imported. Number of records: ', len(df.index),'\n')

# Find unique environments
envs = df.environment_info.unique()
n_envs = len(envs)
print(f'Unique environments: {n_envs}\n {envs}')
````

````python
# Filter rows based on environment
n = 5
print(f'Environment: {envs[n]}')
subset = df[df['environment_info'] == envs[n]]
````

````python
def error_plot(x, y, y_err, title, y_label, x_label):
    PLOT_H = 2
    PLOT_W = 6
    TIME_STEP_TICK = 1
    X_LOW = -0.25
    X_UPPER = len(x)
    Y_LOW = 0.0
    Y_UPPER = 1.1
    Y_GRID = 0.25

    fig, ax = plt.subplots(1, 1, figsize=(PLOT_W, PLOT_H), dpi= 80)
    ax.errorbar(x, y, yerr=y_err, fmt='-o')
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)
    ax.set_title(title, fontsize=12)
    ax.set(ylim=[Y_LOW, Y_UPPER])
    ax.set_xticks(np.arange(0, X_UPPER, TIME_STEP_TICK))
    ax.set_xticklabels(x[::TIME_STEP_TICK], fontdict={'fontsize':10})
    plt.yticks(np.arange(Y_GRID, Y_UPPER, Y_GRID), fontsize=10)
    plt.xlim(X_LOW, X_UPPER)

    # Draw Tick lines  
    for y in np.arange(Y_GRID, Y_UPPER, Y_GRID):    
        plt.hlines(y, xmin=X_LOW, xmax=X_UPPER+1, colors='black', alpha=0.3, linestyles="--", lw=0.5)

    # Lighten borders
    plt.gca().spines["top"].set_alpha(0)
    plt.gca().spines["bottom"].set_alpha(.3)
    plt.gca().spines["right"].set_alpha(0)
    plt.gca().spines["left"].set_alpha(.3)
    plt.show()
````

````python
# Create x and y and y_error
y = subset['RF_Pr']
y_err = subset['RF_Pr_sd']
x = [i for i in range(len(y))]

x_label = 'Training round'

y_label = 'Precision'
title = f'{envs[n]}: REINFORCE {y_label}'

error_plot(x, y, y_err, title, y_label, x_label)
````

````python
# Create x and y and y_error
y = subset['RF_Pr']
y_err = subset['RF_Pr_sd']
x = [i for i in range(len(y))]

# PLOT_H = 4
# PLOT_W = 10
PLOT_H = 2
PLOT_W = 6
TIME_STEP_TICK = 1
X_LOW = -0.25
X_UPPER = len(x)
Y_LOW = 0.0
Y_UPPER = 1.1
Y_GRID = 0.25

fig, ax = plt.subplots(1, 1, figsize=(PLOT_W, PLOT_H), dpi= 80)
ax.errorbar(x, y, yerr=y_err, fmt='-o')
ax.set_xlabel('Training round')
ax.set_ylabel('Precision')
ax.set_title('REINFORCE - Precision with error bars', fontsize=12)
ax.set(ylim=[Y_LOW, Y_UPPER])
ax.set_xticks(np.arange(0, X_UPPER, TIME_STEP_TICK))
ax.set_xticklabels(x[::TIME_STEP_TICK], fontdict={'fontsize':10})
plt.yticks(np.arange(Y_GRID, Y_UPPER, Y_GRID), fontsize=10)
plt.xlim(X_LOW, X_UPPER)

# Draw Tick lines  
for y in np.arange(Y_GRID, Y_UPPER, Y_GRID):    
    plt.hlines(y, xmin=X_LOW, xmax=X_UPPER+1, colors='black', alpha=0.3, linestyles="--", lw=0.5)

# Lighten borders
plt.gca().spines["top"].set_alpha(0)
plt.gca().spines["bottom"].set_alpha(.3)
plt.gca().spines["right"].set_alpha(0)
plt.gca().spines["left"].set_alpha(.3)
plt.show()
````

## Area plots

````python
# Import Data
df = pd.read_csv('bearing_failure_data.csv')
df = df[['time', 'vibration', 'temperature']]
print('bearing failure data imported. Number of records: ', len(df.index),'\n')
normalized_df = (df-df.min())/(df.max()-df.min())
df = normalized_df
df.head()
````

````python
# Prepare Data
x = df['time'].values.tolist()
y1 = df['vibration'].values.tolist()
y2 = df['temperature'].values.tolist()
mycolors = ['tab:red', 'tab:blue', 'tab:green', 'tab:orange', 'tab:brown', 'tab:grey', 'tab:pink', 'tab:olive']      
columns = ['vibration', 'temperature']

# Draw Plot 
fig, ax = plt.subplots(1, 1, figsize=(10,4), dpi= 80)
ax.fill_between(x, y1=y1, y2=0, label=columns[1], alpha=0.5, color=mycolors[1], linewidth=2)
ax.fill_between(x, y1=y2, y2=0, label=columns[0], alpha=0.5, color=mycolors[0], linewidth=2)

# Decorations
ax.set_title('Bearing Failure data', fontsize=18)
ax.set(ylim=[0.0, 1.2])
ax.legend(loc='best', fontsize=12)
plt.xticks(x[::50], fontsize=10, horizontalalignment='center')
plt.yticks(np.arange(0.25, 1.20, 0.25), fontsize=10)
plt.xlim(0.0, x[-1])

# Draw Tick lines  
for y in np.arange(0.25, 1.20, 0.25):    
    plt.hlines(y, xmin=0, xmax=len(x), colors='black', alpha=0.3, linestyles="--", lw=0.5)

# Lighten borders
plt.gca().spines["top"].set_alpha(0)
plt.gca().spines["bottom"].set_alpha(.3)
plt.gca().spines["right"].set_alpha(0)
plt.gca().spines["left"].set_alpha(.3)
plt.show()
````

````python
# Prepare Data
df = pd.read_csv('bearing_failure_data.csv')
df = df[['t_step', 'vibration', 'temperature']]
n_records = len(df.index)
print('bearing failure data imported. Number of records: ', n_records,'\n')

x = df['t_step'].values.tolist()
y1 = df['vibration'].values.tolist()
y2 = df['temperature'].values.tolist()
#mycolors = ['tab:red', 'tab:blue', 'tab:green', 'tab:orange', 'tab:brown', 'tab:grey', 'tab:pink', 'tab:olive']      
#columns = ['vibration', 'temperature']

TIME_STEP_TICK = 5 # X axes, tick marks
    
# Plot Line1 (Left Y Axis)
fig, ax1 = plt.subplots(1,1,figsize=(10, 4), dpi= 80)
ax1.plot(x, y1, color='tab:red')

# Plot Line2 (Right Y Axis)
ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
ax2.plot(x, y2, color='tab:blue')

# Decorations
# ax1 (left Y axis)
ax1.set_xlabel('Time', fontsize=16)
ax1.tick_params(axis='x', rotation=0, labelsize=10)
ax1.set_ylabel('Vibration', color='tab:red', fontsize=16)
ax1.tick_params(axis='y', rotation=0, labelcolor='tab:red' )
ax1.grid(alpha=.4)

# ax2 (right Y axis)
ax2.set_ylabel("Temperature", color='tab:blue', fontsize=16)
ax2.tick_params(axis='y', labelcolor='tab:blue')
ax2.set_xticks(np.arange(0, len(x), TIME_STEP_TICK))
ax2.set_xticklabels(x[::TIME_STEP_TICK], rotation=90, fontdict={'fontsize':10})
ax2.set_title("Bearing failure data", fontsize=18)
fig.tight_layout()
plt.show()
````
